**The Decision Framework**

* **The Search Question:** Which published pages are actively entering a downward traffic trajectory and offer the highest potential recovery if prioritized for an editorial refresh this week?
* **The Unit of Analysis:** The individual web page (`content_id` within a specific `client_id`).
* **The Model Output:** A priority probability score for each page, surfaced as an automated **Top-50 Refresh Queue**.
* **The Action Taken:** A content manager or editor takes this queue and assigns their limited weekly writing bandwidth strictly to updating the top-ranked pages on the list.
* **The Cost of a Wrong Call:**
    * **False Positives (Wasted Opex):** My model incorrectly flags a healthy page as decaying. The business pays a writer to spend hours rewriting content that did not need fixing, wasting valuable budget and time.
    * **False Negatives (Missed Revenue):** My model fails to catch a high-value landing page that is actively losing page-1 visibility on Google. The page continues to bleed organic traffic and conversions to competitors before anyone catches it manually.

**Why Machine Learning is required here (vs. simple rules):**
A basic heuristic like *"just refresh any page where traffic dropped 10%"* fails in the real world. A page might drop in impressions simply because of seasonal demand dips (not true content decay), or it might have massive impressions but a broken meta-title requiring a 5-minute fix rather than a full rewrite. Machine learning is necessary to untangle these multi-signal relationships across 30,000 pages to isolate real structural obsolescence.

**Empirical Justification for Lane 2 (Content Refresh Prioritization)**

To prove why building a Machine Learning model is necessary for prioritizing content updates and why simple heuristics fail. I loaded the 30,000-row starter dataset (`content_refresh_anonymized.csv`) and analyzed three critical baseline statistics:

1. **The Search Volume Fallacy (r = 0.0012):** The correlation between a keyword's generic search volume and a page's actual 90-day impressions is virtually zero. This proves we cannot simply tell writers to "target bigger keywords." We must model historical performance data to see what is actually driving traffic.
2. **The Position CTR Cliff (0.49% vs. 0.04%):** Click-through rates drop drastically from Top-3 rankings (around 0.49%) down to deep pages (around 0.04%). Catching a page as it slips from Position 3 to Position 6 is critical to preventing massive organic revenue loss.
3. **The Word Count Fallacy (2,850 vs. 2,910 words):** Pages with a growing traffic trend average approx. 2,850 words, while declining pages average approx. 2,910 words. Simply "writing longer articles" does not prevent content decay. We need algorithmic feature weighting to spot real structural signals.

```python
import pandas as pd

# 1. Load the FlyRank starter dataset
# (Using the standard relative path from the work/notebooks/ folder)
# Pull the dataset directly from your GitHub repository into Colab:
!git clone https://github.com/Shakir-j/flyrank-internship-ml.git
df = pd.read_csv("flyrank-internship-ml/data/raw/content_refresh_anonymized.csv")

print(f"Starter Dataset Loaded: {df.shape[0]:,} rows across {df['client_id'].nunique()} unique clients.")
print("-" * 60)
print()


# 2. Number 1: Search Volume vs. Actual Impressions Correlation
vol_imp_corr = df['search_volume'].corr(df['impressions_90d'])

print(f"1. Search Volume vs. 90-Day Impressions Correlation: {vol_imp_corr:.4f}")
print()


# 3. Number 2: CTR Drop-off by Position Bucket
if 'position_bucket' in df.columns and 'ctr' in df.columns:
    ctr_by_pos = df.groupby('position_bucket')['ctr'].mean() * 100

    print("2. Mean CTR by Position Bucket:")
    for bucket, ctr_val in ctr_by_pos.items():
        print(f"   - {bucket}: {ctr_val:.2f}%")
    print()


# 4. Number 3: Word Count Parity (Growing vs. Declining Pages)
if 'trend_direction' in df.columns and 'word_count' in df.columns:
    words_by_trend = df.groupby('trend_direction')['word_count'].mean()

    print("3. Mean Word Count by Trend Direction:")
    for trend, words in words_by_trend.items():
        print(f"   - {trend}: {words:.0f} words")
    print()
```
